```python
import pandas as pd
import numpy as np
```

```python
# read data
weather_data = pd.read_csv("jena_climate_2009_2016.csv")
weather_data.head()
```

```python
# Convert Date time column to Datetime64 and make it index
weather_data.index = weather_data['Date Time'].astype('datetime64')
weather_data.drop(columns=['Date Time'],inplace=True)
weather_data.head()
```

```python
# then Resample it to make it hourly
weather_data = weather_data.resample('H').first()
weather_data.head()
```

# Following is data windowing

```python
""" Here slices for windowing are created """
l = 30
m = 3
n = 5

np_data = np.array(weather_data)  # Convert to the numpy array for slicing data
N = np_data.shape[0] # Total number of samples
k = N - (l + m + n)

# Create Input and output Slice
in_slice = np.array([range(i, i + l) for i in range(k)])
op_slice = np.array([range(i + l + m, i + l + m + n) for i in range(k)])
```

```python
op_slice
```

```python
in_data = np_data[in_slice,:]
in_data.shape
```

```python
out_data = np_data[op_slice,:]
out_data.shape
```
